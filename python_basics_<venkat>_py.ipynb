```python
age_in_months = age * 12
age_in_days = age * 365
remainder = age % 7
age_power_2 = age ** 2

print("Age in months:", age_in_months)
print("Age in days:", age_in_days)
print("Remainder when age is divided by 7:", remainder)
print("Age raised to the power of 2:", age_power_2)
```
